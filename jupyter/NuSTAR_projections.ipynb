```python
def theta2tau(s2theta, ms):
    """ Convert from ms [keV] and sin2theta to lifetime [s]
    """
    
    return 1./(1.361e-29*(s2theta/1.e-7)*ms**5.)

hbar = 6.582119569e-16 # [eV.s]

def theta2gagg(s2theta, ms):
    """ Convert from ms [keV] and sin2theta to gagg [1/GeV]
    """
    
    Gamma = hbar/theta2tau(s2theta, ms)/2. # [eV]
    
    return np.sqrt(64.*np.pi*Gamma/(ms*1.e3)**3.)*1.e9 # [1/GeV]
```

# NuSTAR Projection Estimates

```python
#read in https://arxiv.org/abs/2207.04572
#note this paper uses data up to 2017 but has two complementary detection strategies and more aggressive filtering

nustar_sin2_beacom = np.loadtxt('../existing_limits/NuSTAR_sin2theta_Beacom.csv', delimiter=',')

plt.plot(nustar_sin2_beacom[:,0], nustar_sin2_beacom[:,1])
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$m_{\chi}$ [keV]')
plt.ylabel(r'$\sin^2(2\theta)$')

plt.show()
```

```python
#convert to tau, and x2 for two photon decay

mass_keV = nustar_sin2_beacom[:,0]
tau_s = 2.0 * theta2tau(nustar_sin2_beacom[:,1], mass_keV)

plt.plot(mass_keV, tau_s)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$m_{\chi}$ [keV]')
plt.ylabel(r'$\tau$ [s]')

plt.show()
```

```python
#now let's scale by exposure time since the 2017 data they used

#they used 7 Ms * 2 and then 20 Ms * 8 for their combined joint techniques

#let's estimate that the amount of blank sky data scales with years

#2012-2017 is 5 years
#2018-2025 is 8 years

#let's also say take more data: they do b > 15, we can do say b > 2, so factor sqrt(15/2)

exp_fac = (8.0+5.0)/5.0
print(exp_fac)
print(np.sqrt(exp_fac))

new_tau_s_raw = tau_s * np.sqrt(exp_fac) * np.sqrt(15.0/2.0) #improvement in limit by sqrt(exp time) bc S/sqrt(B)


plt.plot(mass_keV, new_tau_s_raw)
plt.plot(mass_keV, tau_s, linestyle='dashed')

#plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$m_{\chi}$ [keV]')
plt.ylabel(r'$\tau$ [s]')
plt.ylim([1e27, 1e30])

plt.show()
```

```python
#smooth

#let's smooth this 

new_tau_s = savgol_filter(new_tau_s_raw, window_length=80, polyorder=3)

plt.plot(mass_keV, new_tau_s, color='k')
plt.plot(mass_keV, new_tau_s_raw, color='gray', linestyle='dotted')
plt.plot(mass_keV, tau_s, linestyle='dashed')

#plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$m_{\chi}$ [keV]')
plt.ylabel(r'$\tau$ [s]')
plt.ylim([1e27, 1e30])

plt.show()
```
